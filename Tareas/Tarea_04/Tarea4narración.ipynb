Primero importamos la librería pandas y cargamos la base de datos, identificando el separador con en que venía. Finalmente visualizamos las primeras  filas de la base de datos.

```python
import pandas as pd

archivo1 = pd.read_csv('ultima.csv',delimiter=";")
archivo1.head()
```

Luego importamos una librería para poder realizar un gráfico, definimos las variables en los ejes y lo visualizamos.

```python
import matplotlib.pyplot as plt
años = archivo1['ano']
cantidad = archivo1['total']
plt.figure(figsize=(8,5))
plt.plot(años, cantidad, marker='o', linestyle='-')
plt.title("Drogas por año")
plt.ylabel("Cantidad")
plt.xlabel("Año")
plt.show()

```
